# run all plays

```python
import multiprocessing
from tqdm import tqdm
```

```python
for pid in tqdm(record_df.index, total=len(record_df.index)):
    play = get_play(pid)
    events = play.events
    marks = []
    for id_, _ in enumerate(events):
        event_time = events[id_][0]
        hit_type = transform_drum_note_hit_type(events[id_][1])
        if hit_type > 0:
            marks.append((event_time, COLORS[hit_type]))

    figname = str(pid) + '-acc.png'
    plot_raw_acc_signal(play.play_dict['L'], play.play_dict['R'], marks=marks, figname=figname)
    figname = str(pid) + '-gyr.png'
    plot_raw_gyr_signal(play.play_dict['L'], play.play_dict['R'], marks=marks, figname=figname)
```
